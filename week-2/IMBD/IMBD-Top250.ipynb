```python
# Importing necessary Libraries

import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
htmlPage = requests.get("https://www.imdb.com/chart/top/")
```

```python
soup = BeautifulSoup(htmlPage.content,"html.parser")
soup
```

```python
index = []
title = []
year = []
rating = []
```

```python
titleSoup = soup.find_all('td',attrs={
    'class':"titleColumn"
})
```

```python
print(titleSoup)
```

```python
for oneTitle in titleSoup:
    # print(oneTitle.text.strip())
    # print(len(oneTitle.text.strip()))
    temp = oneTitle.text.strip().split()
    titleTemp = ""
    for j in temp:
        j = j.replace(".","")
        print(j)
        if j.isdigit():
            index.append(j)
        elif j.isalpha():
            if j != "":
                titleTemp += j
                titleTemp += " "
        else:
            j = j.replace("(","").replace(")","")
            if len(j)==4:
                year.append(j)
    if len(titleTemp) != 0:
        title.append(titleTemp[:-1])
```

```python
print(index)
```

```python
print(title)
```

```python
print(year)
```

```python
rating = soup.find_all('td',attrs={
    'class':"ratingColumn imdbRating"
})

ratings = []
for oneTitle in rating:
    ratings.append(oneTitle.text.strip())
```

```python
df = {
    'S.No':index[:244],
    "Title":title[:244],
    "Year":year[:244],
    "Ratings":ratings[:244]

}

df = pd.DataFrame(df)
```

```python
df.to_csv("IMBD-Top250_Movies.csv",index=False)
```
